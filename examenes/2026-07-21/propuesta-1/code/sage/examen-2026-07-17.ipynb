```python
# Define the matrix A
A = matrix(QQ, [[-4, 0, -1, 0], [0, -2, 0, 0], [4, 1, 0, 0], [6, 1, 3, -2]])

# Compute the Jordan form (J) and the transformation matrix (P)
J, P = A.jordan_form(transformation=True)

print("Jordan Form J:")
print(J)
print("\nTransformation Matrix P:")
print(P)
```

```python
# Check if A equals P * J * P^-1
print(A == P * J * P^(-1))
```
