```julia
using Plots
using LaTeXStrings
```

```julia
# Potencial sobre la superficie del cilindro
V = 1.0

f(θ) = sign(sin(2θ))*V

θ = range(0, 2π, length=2000)

plot(
    θ,
    f.(θ),
    lw=3,
    xlabel=L"\theta",
    ylabel=L"\Phi(b,\theta)",
    xticks=([0, π/2, π, 3π/2, 2π],
            [L"0", L"\pi/2", L"\pi", L"3\pi/2", L"2\pi"]),
    yticks=([-V, V], [L"-V", L"V"]),
    legend=false,
    xlims=(0,2π),
    ylims=(-1.3V,1.3V),
    framestyle=:box,
    dpi=300
)
```

```julia
V = 1.0
b = 1.0

# Número de puntos
N = 800

θ = range(0, 2π, length=N)

x = b .* cos.(θ)
y = b .* sin.(θ)

# Potencial sobre el borde
ϕ = V .* sign.(sin.(2 .* θ))

# Colores según el signo del potencial
colors = ifelse.(ϕ .> 0, :red, :blue)

scatter(
    x,
    y,
    markercolor = colors,
    markersize = 5,
    markerstrokewidth = 0,
    aspect_ratio = :equal,
    legend = false,
    xlabel = L"x",
    ylabel = L"y",
    xlims = (-1.2b,1.2b),
    ylims = (-1.2b,1.2b),
    framestyle = :box,
    dpi = 300,
)

# Dibujar la circunferencia
plot!(
    b*cos.(θ),
    b*sin.(θ),
    color = :black,
    lw = 2,
)
```

```julia
b = 1.0

p = plot(
    aspect_ratio = :equal,
    legend = false,
    xlabel = L"x",
    ylabel = L"y",
    xlims = (-1.2, 1.2),
    ylims = (-1.2, 1.2),
    framestyle = :box,
    dpi = 300,
)

# First quadrant (+V₀)
θ = range(0, π/2, length=200)
plot!(p, b*cos.(θ), b*sin.(θ), lw=6, color=:red)

# Second quadrant (-V₀)
θ = range(π/2, π, length=200)
plot!(p, b*cos.(θ), b*sin.(θ), lw=6, color=:blue)

# Third quadrant (+V₀)
θ = range(π, 3π/2, length=200)
plot!(p, b*cos.(θ), b*sin.(θ), lw=6, color=:red)

# Fourth quadrant (-V₀)
θ = range(3π/2, 2π, length=200)
plot!(p, b*cos.(θ), b*sin.(θ), lw=6, color=:blue)

# Coordinate axes
plot!(p, [-1.2, 1.2], [0, 0], color=:black, lw=1)
plot!(p, [0, 0], [-1.2, 1.2], color=:black, lw=1)

# Labels
annotate!( 0.60,  0.60, text(L"V", 14, :red))
annotate!(-0.60,  0.60, text(L"-V", 14, :blue))
annotate!(-0.60, -0.60, text(L"V", 14, :red))
annotate!( 0.60, -0.60, text(L"-V", 14, :blue))

display(p)
```
